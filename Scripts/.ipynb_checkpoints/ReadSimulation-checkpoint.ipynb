```python
class ResultsSim(object):
    def __init__(self, file_name):
        self.file_name = file_name
        with gzip.open(file_name, 'rb') as rs:
            self.data_dict = pickle.load(rs)
        
    def num_simulations(self,k1,k2):
        return len(self.data_dict[k1][k2].keys())-1
    
    def get_inputs(self,k1,k2):
        return self.data_dict[k1][k2]['inputs']
        
    def mink(self,k1,k2):
        return self.data_dict[k1][k2]['inputs']['mink']
        
    def maxk(self,k1,k2):
        return self.data_dict[k1][k2]['inputs']['maxk']
    
    def ResultsKeys(self,k1,k2):
        return self.data_dict[k1][k2]
    
    def eval_coph(self,k1,k2):
        numberSim=self.num_simulations(k1,k2)
        mink=self.mink(k1,k2)
        maxk=self.maxk(k1,k2)
        CophWin={}
        for k in range(mink,maxk):
            CophWin[k]=0
        results=self.data_dict[k1][k2]
            
        for i in range(numberSim):
            kCoph=2
            maxCoph=results[i][2]['coph_corr']
            evarCoph=results[i][2]['evar']
            for k in range(mink+1,maxk):
                if  round(results[i][k]['coph_corr'].item()-maxCoph,3)>0:
                    kCoph=k
                    maxCoph=results[i][k]['coph_corr']
                    evarCoph=results[i][k]['evar']
                if  abs(maxCoph-results[i][k]['coph_corr'].item())<0.001 and results[i][k]['evar']/evarCoph>1.07:
                    kCoph=k
                    maxCoph=round(results[i][k]['coph_corr'].item(),3)
                    evarCoph=results[i][k]['evar']    
            CophWin[kCoph]+=1
        return CophWin

    def eval_silh(self,k1,k2):
        numberSim=self.num_simulations(k1,k2)
        mink=self.mink(k1,k2)
        maxk=self.maxk(k1,k2)
        SilWin={}
        for k in range(mink,maxk):
            SilWin[k]=0
        results=self.data_dict[k1][k2]
            
        for i in range(numberSim):
            kSil=2
            maxSil=results[i][2]['silhouette']['AVG']
            evarSil=results[i][2]['evar']

            for k in range(mink+1,maxk):
                if round(results[i][k]['silhouette']['AVG']-maxSil,3)>0:
                    kSil=k
                    maxSil=results[i][k]['silhouette']['AVG']
                    evarSil=results[i][k]['evar']
                if abs(results[i][k]['silhouette']['AVG']-maxSil)<0.001 and results[i][k]['evar']/evarSil>1.07:
                    kSil=k
                    maxSil=results[i][k]['silhouette']['AVG']
                    evarSil=results[i][k]['evar']
            SilWin[kSil]+=1
        return SilWin
        
    def eval_silh_local_max(self,k1,k2):
        numberSim=self.num_simulations(k1,k2)
        mink=self.mink(k1,k2)
        maxk=self.maxk(k1,k2)
        SilWin={}
        for k in range(mink,maxk):
            SilWin[k]=0
        results=self.data_dict[k1][k2]    
        for i in range(numberSim):
            for k in range(mink+1,maxk-1):
                if results[i][k]['silhouette']['AVG']>results[i][k-1]['silhouette']['AVG'] and results[i][k]['silhouette']['AVG']>results[i][k+1]['silhouette']['AVG']:
                    SilWin[k]+=1
        return SilWin
    
    def eval_coph_local_max(self,k1,k2):
        numberSim=self.num_simulations(k1,k2)
        mink=self.mink(k1,k2)
        maxk=self.maxk(k1,k2)
        SilWin={}
        for k in range(mink,maxk):
            SilWin[k]=0
        results=self.data_dict[k1][k2]    
        for i in range(numberSim):
            for k in range(mink+1,maxk-1):
                if results[i][k]['coph_corr']>results[i][k-1]['coph_corr'] and results[i][k]['coph_corr']>results[i][k+1]['coph_corr']:
                    SilWin[k]+=1
        return SilWin
    
    def distance_original(self,dist,k1,k2):
        k=self.data_dict[k1][k2]['inputs']['n_processes']
        results=self.data_dict[k1][k2]
        numSim=self.num_simulations(k1,k2)
        AVGdist=0
        for i in range(numSim):
            FP=results[i][k]['processes']
            FP[FP<0]=0
            P=results[i]['TrueProcesses']
            AVGdist+=calculate_dist(FP.T,P,dist)
        return AVGdist/numSim/k

    def plot_signatures(self,nsim,dist,k1,k2):      
        results=self.data_dict[k1][k2]
        k=results['inputs']['n_processes']
        FP=results[nsim][k]['processes']
        FP[FP<0]=0
        P=results[nsim]['TrueProcesses']
        for i in range(k):
            plot_sign(FP,P,i,dist)
            
    
    def metrics_k1(self,k1,dist,typeEval):
        # Type evaluation = "Max" (Maximum metric)
        # Type evaluation = "LMax" (Local Maximum metric)
        print("Results for k1 = "+str(k1))
        results=self.data_dict[k1]
        df = pd.DataFrame(np.zeros((len(results.keys()),3)),index=results.keys(), columns=["%Sil","%Coph","Dist"])
        if typeEval == "Max":
            for k2 in results.keys():
                if k2 < 1:
                    p=k1
                else:
                    p=k2
                sil=self.eval_silh(k1,k2)
                df.loc[k2,"%Sil"]=sil[p]/sum(sil.values())
                coph=self.eval_coph(k1,k2)
                df.loc[k2,"%Coph"]=coph[p]/sum(coph.values())
                df.loc[k2,"Dist"]=self.distance_original(dist,k1,k2)
        else:
            for k2 in results.keys():
                if k2 < 1:
                    p=k1
                else:
                    p=k2
                sil=self.eval_silh_local_max(k1,k2)
                df.loc[k2,"%Sil"]=sil[p]/sum(sil.values())
                coph=self.eval_coph_local_max(k1,k2)
                df.loc[k2,"%Coph"]=coph[p]/sum(coph.values())
                df.loc[k2,"Dist"]=self.distance_original(dist,k1,k2)            
        return(df)
    
    def metrics(self,dist,typeEval):
        results=self.data_dict
        for k1 in results.keys():
            print(self.metrics_k1(k1,dist,typeEval))
        
```
